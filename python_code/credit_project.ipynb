```python
X = df.drop(columns=['Cost_Matrix'])
```

```python
y=df['Cost_Matrix']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 100)
```

```python
clf.fit(X_train, y_train)
```

```python
importance = clf.feature_importances_
```

```python
for feature, score in zip(df.columns, importance):
    print('Feature: %s, Score: %.5f' % (feature, score))
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(importance)), importance)
plt.xticks(range(len(importance)), [f"Feature: {feature}" for feature, score in zip(df.columns, importance)], rotation=90)
plt.show()
```

```python
from sklearn import metrics
y_pred = clf.predict(X_test)
```

```python
print("ACCURACY OF THE MODEL:", metrics.accuracy_score(y_test, y_pred))
```

```python
#accuracy is not good lets try to make it good.
#dataset is imbalanced lets make it balanced.
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

smote = SMOTE()

# Apply SMOTE to generate synthetic samples
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert the resampled data back to DataFrame and Series if needed
X = pd.DataFrame(X_resampled, columns=X.columns)
y = pd.Series(y_resampled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators = 100)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print("ACCURACY OF THE MODEL:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#Accuracy increased by 6% after SMOTE
```
